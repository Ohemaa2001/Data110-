```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns

iris = sns.load_dataset('iris')


iris_clean = iris.dropna()

features = ['petal_width','petal_length']

scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris_clean[features])

kmeans = KMeans(n_clusters=3, random_state=42)
iris_clean['cluster'] = kmeans.fit_predict(iris_scaled)

plt.scatter(iris_clean['sepal_length'], iris_clean['petal_width'], c=iris_clean['cluster'], cmap='viridis')

plt.xlabel('(sepal_length)')
plt.ylabel('sepal_width (mm)')
plt.title('K-Means Clustering of Iris')

plt.show()
```
