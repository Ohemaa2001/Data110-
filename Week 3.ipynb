# New Section

```
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 20, 200)
plt.plot(x, np.sin(x))
plt.show()
```

```
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.plot(x, y)
plt.title("Quadratic Function")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```
x = np.random.rand(90)
y = np.random.rand(90)
plt.scatter(x, y)
plt.title("Random Scatter Plot")
```

```
plt.plot(x, y, color='purple', linestyle='--', linewidth=2, marker='o')
```

```
x = np.linspace(-12, 14, 200)
plt.plot(x, x**2, x, x**3, x, np.sin(x))
plt.legend(['x^2', 'x^3', 'sin(x)'])
```

```
x = np.linspace(-20, 20, 100)
plt.plot(x, np.tan(x))
plt.title("Tangent Function")
plt.xlim(-200, 200)
plt.ylim(-100, 100)
```
