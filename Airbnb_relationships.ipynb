```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Ohemaa2001/Data110-/refs/heads/main/airbnb_ny19.csv")
df.head()
df.info()


x = [1, 2, 3, 4]
y = [10, 8, 12, 16]

plt.bar(x, y, label='Price by Location')
plt.legend()
plt.show()

```

```python
#bar graph for room_type



# Assuming 'room_type' is a column in your DataFrame 'df'
room_type_counts = df['room_type'].value_counts()

plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.bar(room_type_counts.index, room_type_counts.values)
plt.xlabel("Room Type")
plt.ylabel("Number of Listings")
plt.title("Distribution of Room Types")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# prompt: any histogram from this dataset?

# Histogram for price
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=30, edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show()
```

```python
# prompt: group by neighbourhood_group and then do bar graph

# Group data by 'neighbourhood_group' and count occurrences
neighbourhood_group_counts = df.groupby('neighbourhood_group')['id'].count()

# Create a bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.bar(neighbourhood_group_counts.index, neighbourhood_group_counts.values)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Number of Listings')
plt.title('Number of AirBnB Listings per Neighbourhood Group')
plt.show()
```
