```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Emergency_Departent_Visits_df=pd.read_csv('https://raw.githubusercontent.com/Ohemaa2001/Data110-/refs/heads/main/Empty%20Folder/Emergency_Department_Visits_in_the_United_States_from_2016-2020.csv')
Emergency_Departent_Visits_df.describe()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.hist(Emergency_Departent_Visits_df['Group'], bins=15, color='skyblue',edgecolor='white')
plt.title('Distribution of Groups Admitted to Emergency Room')
plt.xlabel('Groups')
plt.ylabel('Estimate')
plt.grid(axis='x')
plt.xticks([0, 5])
plt.tick_params(axis='x', colors='gray')  # Set x-axis tick colors to gray
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10, 6))
plt.hist(Emergency_Departent_Visits_df['Measure'], bins=15, color='skyblue',edgecolor='white')
plt.title('Distribution of Diagnosis Admitted to Emergency Room')
plt.xlabel('Measure')
plt.ylabel('Estimate')
plt.grid(axis='x')
plt.xticks([0, 40])
plt.tick_params(axis='x', colors='gray')  # Set x-axis tick colors to gray
plt.show()
```
