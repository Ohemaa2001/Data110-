```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/Ohemaa2001/Data110-/refs/heads/main/Empty%20Folder/Emergency_Department_Visits_in_the_United_States_from_2016-2020.csv')

aggregated_data = data.groupby('Year')['Estimate'].mean().reset_index()

x = aggregated_data["Year"]
y = aggregated_data["Estimate"]

plt.plot(x, y, marker="o", label= "Visits With the Years" )


plt.plot(x, y, marker="o", label= "Visits With the Years" )
plt.xlabel("Year")
plt.ylabel("Estimate")
plt.title("Emergency Department Visits in the United States from 2016 to beginning of 2020")
plt.legend()
plt.grid(True)
plt.show()
```
